````python
import os
from PIL import Image
import numpy as np

def load_xray_dataset(dataset_dir):
   
    images = []
    labels = []

    # Define label mapping: 0 for normal, 1 for ill
    label_mapping = {"normal": 0, "pneumonia": 1}

    for label in label_mapping.keys():
        label_dir = os.path.join(dataset_dir, label)
        if not os.path.exists(label_dir):
            continue

        for filename in os.listdir(label_dir):
            img_path = os.path.join(label_dir, filename)
            try:
                # Attempt to open the image file with PIL
                img = Image.open(img_path)
                #img.verify()  # Verify the image file's integrity
               # img = np.array(img)  # Convert PIL image to NumPy array
                img.resize((128,128))
                images.append(img)
                labels.append(label_mapping[label])
            except (IOError, SyntaxError) as e:
                # If the image is corrupted, print an error message and continue
                print(f"Removing {img_path} due to error: {e}")
                os.remove(img_path)

    return images,labels

# Usage example:
train_dataset_dir = "C://Users/ZBook/Downloads/Tp1/Dataset/chest_xray/train/"
images, labels = load_xray_dataset(train_dataset_dir)

````

## Exploratory Data Analysis (EDA)

### Objective

Understand the dataset's basic statistics and visualize the data.

### Guidance

1. Use matplotlib to visualize some sample images from each class.
2. Investigate the distribution of classes (Pneumonia/Normal).

### Exercise

Perform exploratory data analysis on the dataset. Write your code in the cell below.


````python
import numpy as np
import matplotlib.pyplot as plt

# Perform basic statistical analyses
samples_names =  ["NORMAL", "PNEUMONIA"]
num_samples = len(labels)
num_normal = np.sum(np.array(labels) == 0)
num_pneumonia = np.sum(np.array(labels) == 1)
````

````python
print(f"Total number of samples: {num_samples}")
print(f"Number of normal X-rays: {num_normal}")
print(f"Number of pneumonia X-rays: {num_pneumonia}")

num_samples = 2
num_classes = len(samples_names)
fig, axes = plt.subplots(num_samples, num_classes, figsize=(15, 5))

for i, ax_row in enumerate(axes):
    for j, ax in enumerate(ax_row):
        class_images = [images[k] for k in range(len(images)) if labels[k] == j]
        ax.imshow(class_images[i])
        ax.set_title(samples_names[j])
        ax.axis('off')

plt.show()
plt.show()

````

````python
# Count the number of images for each class
class_counts = [sum(1 for label in labels if label == i) for i in range(len(samples_names))]
print("Number of Normal samples:", class_counts[1])
print("Number of Pneumonia samples:", class_counts[0])

# Visualize the distribution of classes (Pneumonia/Normal)
plt.figure(figsize=(10, 6))
plt.bar(samples_names, class_counts,color=['r','b'])
plt.xlabel('Classes')
plt.ylabel('Number of Samples')
plt.title('Class Distribution')
plt.show()

````

````python
# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(class_counts, labels=samples_names, autopct='%1.1f%%', colors=['r', 'b'])
plt.title('Class Distribution')
plt.show()
````

### Objective

Understand how class imbalance affects model performance and learn ways to mitigate it.

### Exercise

Address the class imbalance and evaluate the impact on the model. Write your code in the cell below.


````python

# Your code here for Class Imbalance

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


test_dataset_dir = "C://Users/ZBook/Downloads/Tp1/Dataset/chest_xray/test/"
test_images, test_labels = load_xray_dataset(test_dataset_dir)

train_data = [np.array(img).flatten()[:16384] for img in images]
test_data = [np.array(img).flatten()[:16384] for img in test_images]

# Create a Decision Tree Classifier
decision_tree = DecisionTreeClassifier(random_state=42)


decision_tree.fit(train_data, labels)
````

````python
from sklearn import tree
tree.plot_tree(decision_tree)
````

````python
from sklearn.metrics import accuracy_score

y_pred = decision_tree.predict(test_data)


accuracy = accuracy_score(test_labels, y_pred)
#accuracy = accuracy_score(test_data, y_pred)
precision = precision_score(test_labels, y_pred)
recall = recall_score(test_labels, y_pred)
f1 = f1_score(test_labels, y_pred)

print(f"Decision Tree Accuracy: {accuracy}")
print(f"Decision Tree Precision: {precision}")
print(f"Decision Tree Recall: {recall}")
print(f"Decision Tree F1-Score: {f1}")
````

## Model Building

### Objective

Build and train a deep learning model for image classification.

### Guidance

1. Use a Convolutional Neural Network (CNN) for this task.
2. Experiment with different architectures and hyperparameters.

### Exercise

Build and train your model. Write your code in the cell below.


````python

# Your code here for Model Building

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

train_data = np.array(train_data)
labels = np.array(labels)
test_data = np.array(test_data)
test_labels = np.array(test_labels)

# Normalize the pixel values (assuming your data is grayscale)
train_data = train_data / 255.0  
test_data = test_data / 255.0 
X_train = train_data.reshape(-1, 128, 128, 1)
X_test = test_data.reshape(-1, 128, 128, 1)

input_layer = keras.Input(shape=(128, 128, 1))


conv1 = layers.Conv2D(32, (3, 3), activation='relu')(input_layer)
pool1 = layers.MaxPooling2D((2, 2))(conv1)
conv2 = layers.Conv2D(64, (3, 3), activation='relu')(pool1)
pool2 = layers.MaxPooling2D((2, 2))(conv2)
conv3 = layers.Conv2D(128, (3, 3), activation='relu')(pool2)
pool3 = layers.MaxPooling2D((2, 2))(conv3)


flatten = layers.Flatten()(pool3)


dense1 = layers.Dense(128, activation='relu')(flatten)
output_layer = layers.Dense(1, activation='sigmoid')(dense1)  # Binary classification


model = keras.Model(inputs=input_layer, outputs=output_layer)


model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
````

````python

# Train the model
history = model.fit(X_train, labels, batch_size=32,
                    validation_data=(X_test, test_labels),
                    epochs=20,  # Adjust the number of epochs
                    verbose=1)
````
